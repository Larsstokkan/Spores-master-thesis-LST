```python
import pandas as pd
import plotly.express as px
import plotly.graph_objs as go
import plotly.colors as pc

# Load CSV-file into a pandas DataFrame
df = pd.read_csv("Measurement 3.csv", sep=";")

# Calculate the mean of the replicates for each column
cols_to_avg = ["N1", "NEW189", "NEW18", "NEW1", "OLD189", "OLD18", "OLD1", "S2-1"]
grouped_df = df[cols_to_avg].groupby(df["Cycle"]).mean()

# Define a continuous color scale
color_scale = pc.sequential.Viridis_r

# Create a new empty figure
fig = go.Figure()

# Plot each line separately with a for loop
for col in cols_to_avg:
    fig.add_trace(go.Scatter(x=grouped_df.index, y=grouped_df[col], mode='lines',
                             name=col, line=dict(color=color_scale[len(fig.data)], width=2)))

# Set the title and axis labels
fig.update_layout(title="Reaction over time", xaxis_title="Cycle", yaxis_title="RFU")

# Display the plot using plotly.offline.iplot()
fig.show()
```

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objs as go

# Load CSV-file into a pandas DataFrame
df = pd.read_csv("Measurement 1.csv", sep=";")

# Calculate the mean of the replicates for each column
cols_to_avg = ["Pos", "Neg", "Spores189", "Spores18", "Spores1"]
grouped_df = df[cols_to_avg].groupby(df["Cycle"]).mean()

# Define the colors for each dataset using hex codes
colors = {"Pos": "#fde725", "Neg": "#90d743", "Spores189": "#35b779",
          "Spores18": "#21918c", "Spores1": "#31688e"}

# Create a new empty figure
fig = go.Figure()

# Plot each line separately with a for loop
for col in cols_to_avg:
    if "Spores" in col:
        fig.add_trace(go.Scatter(x=grouped_df.index, y=grouped_df[col], mode='lines',
                                 name=col, line=dict(color=colors[col.split(" ")[0]], width=2)))
    else:
        fig.add_trace(go.Scatter(x=grouped_df.index, y=grouped_df[col], mode='lines',
                                 name=col, line=dict(color=colors[col], width=2)))

# Set the title and axis labels
fig.update_layout(title="Measurement 1", xaxis_title="Cycles", yaxis_title="RFU")

# Display the plot using plotly.ofline.iplot()
fig.show()
```

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objs as go
import plotly.colors as pc

# Load CSV-file into a pandas DataFrame
df = pd.read_csv("Measurement 3.csv", sep=";")

# Calculate the mean of the replicates for each column
cols_to_avg = ["N1", "NEW189", "NEW18", "NEW1", "OLD189", "OLD18", "OLD1", "S2-1"]
grouped_df = df[cols_to_avg].groupby(df["Cycle"]).mean()

# Define the colors for each dataset using hex codes
colors = {"N1": "#fde725", "NEW189": "#2c728e", "NEW18": "#3b528b",
          "NEW1": "#472d7b", "OLD189": "#addc30", "OLD18": "#5ec962",
          "OLD1": "#28ae80", "S2-1": "#21918c"}

# Define the names for each dataset
names = {"N1": "Negative control", "NEW189": "New spores OD 189", "NEW18": "New spores OD 18.9",
         "NEW1": "New spores OD 1.89", "OLD189": "Test spores OD 189", "OLD18": "Test spores OD 18.9",
         "OLD1": "Test spores OD 1.89", "S2-1": "Standard 2"}

# Create a new empty figure
fig = go.Figure()

# Plot each line separately with a for loop
for col in cols_to_avg:
    fig.add_trace(go.Scatter(x=grouped_df.index, y=grouped_df[col], mode='lines',
                             name=names[col], line=dict(color=colors[col], width=2)))

# Set the title and axis labels
fig.update_layout(title="End measurement", xaxis_title="Cycle", yaxis_title="RFU")
fig.update_layout(font=dict(family="Times New Roman", size=15))

# Display the plot using plotly.offline.iplot()
fig.show()
```
